## Notes

The geopotential height data was obtained from:
/archive/wfc/SPEAR/SPEAR_c192_o1_Hist_AllForc_IC1921_K50_ens_01_03/pp_ens_01/atmos/ts/monthly/90yr/atmos.192101-201012.hght.nc

This code just gets the geopotential height at 500hPa

```python
import xarray as xr

variable = "hght"
file_name = "RawData/atmos.192101-201012.hght.nc"
out_name = "DATA/atmos.192101-201012.Z500.nc"

data_file = xr.open_mfdataset(file_name, combine='by_coords', decode_timedelta=True)
hght = data_file[variable]

z500_data = hght.sel(level=500, method='nearest')

z500_data = z500_data.rename("Z500")
z500_data.attrs["long_name"] = "Geopotential Height at the 500 hPa pressure level"
z500_data.attrs["note"] = ""

encoding = {"Z500": {"zlib": True, "_FillValue": hght.encoding.get("_FillValue", None)}}
z500_data.to_netcdf(out_name, encoding=encoding)

```
